```python
# n_amount is similar to n_amount_management. The sum of fertilisation dose/s is equal to n_amount. we can take n_amount by itself.
# similarly, density_relative has a duplicate in management
columns_to_drop = ['n_amount_management',
       'n_fertilisation_number', 'n_fertilisation_dose_1',
       'n_fertilisation_dose_2', 'n_fertilisation_dose_3',
       'n_fertilisation_date_1', 'n_fertilisation_date_2',
       'n_fertilisation_date_3', 'density_relative_management']

df.drop(columns_to_drop, inplace=True, axis=1)
```

```python
def create_features(df):
    # Convert dates
    df['sowing_date'] = pd.to_datetime(df['sowing_date'])
    df['harvest_date'] = pd.to_datetime(df['harvest_date'])
    
    # Create growing season length (critical feature)
    df['growing_days'] = (df['harvest_date'] - df['sowing_date']).dt.days
    
    
    # density metrics
    # Absolute difference from target
    df['density_error'] = df['plant_density'] - df['density_level']
    
    #  Density x Nitrogen interaction
    df['density_n_ratio'] = df['plant_density'] / df['n_amount']
    
    #  Crowding indicator (simplified)
    df['crowding'] = np.where(df['plant_density'] > 1.2*df['density_level'], 1, 0)
    
    # Soil water holding capacity (simplified):
    # Approximated based on relative contributions of clay (0.45), silt (0.35), and sand (0.15),
    # with a penalty for high pH (above 7), reflecting reduced availability in alkaline soils.
    # This is a heuristic model, not based on detailed lab-calibrated field capacity equations.
    df['soil_water_capacity'] = (
        df['soil_clay']*0.45 + 
        df['soil_silt']*0.35 + 
        df['soil_sand']*0.15 -
        df['soil_ph'].apply(lambda x: 0.1*x if x>7 else 0)
    )
    
    # Nitrogen Use Efficiency (NUE)
    df['NUE'] = df['biomass_seed'] / df['n_amount'].replace(0, np.nan) # g biomass per g N
    
    
    # Growing degree days (GDD)
    df['GDD'] = ((df['mean_temperature'] + df['max_temperature'])/2 - 10).clip(lower=0) * df['growing_days']

    return df
```

Feature Engineering:
- growing_days: The number of days from sowing to harvest
- soil_texture: Processed according to USDA triangle
- soil_water_capacity: water capacity calculation based on soil
- NUE: Nitrogen Use Efficiency
- GDD: growing degree days
- RUE: radiation use efficiency
- density_error: absolute difference of plant_density compared to target density
- density_n_ratio: interaction of density and nitrogen
- crowding: indicates how cramped it is

### Numerical Features

```python
## Get the correlation to determine which feature to drop

numerical_cols = df.select_dtypes(include='number').columns.tolist()

correlation_matrix = df[numerical_cols].corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Key insights
- N_amount and N_amount_management is the same. Drop 1 column
- density level is the target plant density and thus are highly correlated. Can possible bin into several groups as categorical variable
- density_relative and density_relative_management should be removed
- plant_density and biomass_seed(yield) is not directly correlated
- growing_days seems to be moderately correlated with yield (0.48)
- higher latitude = higher number of high humidity days
- RUE is highly correlated with biomass seed; NUE and growing_days are moderately correlated
- max_temp, min_temp, and high_heat_stress_days are highly correlated. To avoid multicollinearity, it may be better to keep only 1 of them.

Context-specific decisions:
- Dropping 'year' feature. It does not contribute much meaning.
- Keeping heat_stress_days and dropping max_temp and mean_temp.
- Drop target density, as plant density is much important for the task we have.
- plot_lat is to be removed as high_humidity_days suggest the same.
- dropping mean_pet as it is highly correlated with heat_stress_days which is more relevant.

```python
addtl_columns_todrop = [
    'year', 'max_temperature', 'mean_temperature',
    'density_level', 'plot_lat', 'mean_pet'
]

df.drop(addtl_columns_todrop, inplace=True, axis=1, errors='ignore')
```
